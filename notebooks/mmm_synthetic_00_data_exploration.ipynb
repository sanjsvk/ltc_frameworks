# 00 — Synthetic Data Exploration

Load and visualise all 5 scenarios. Validate ground-truth decomposition.

```python
import pandas as pd
import matplotlib.pyplot as plt
from ltc.data import load_all_scenarios, split_observed_truth
from ltc.data.loader import SCENARIOS
```

```python
scenarios = load_all_scenarios("data/raw")
for name, df in scenarios.items():
    print(f"{name}: {df.shape}, net_sales mean={df.net_sales_observed.mean():.2f}")
```

```python
# Validate S2 spend pause (weeks 104-112)
df_s2 = scenarios["S2"]
fig, ax = plt.subplots(figsize=(14,4))
ax.plot(df_s2["spend_tv"], label="TV Spend")
ax.axvspan(104, 112, alpha=0.2, color="red", label="Spend Pause")
ax.legend()
plt.title("S2: TV Spend with Pause Window")
plt.show()
```
